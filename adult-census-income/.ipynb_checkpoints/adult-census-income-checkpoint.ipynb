# Data Preprocessing 

```python
def data_preprocessing(X, y, numeric_features, categorical_features, reSampling:bool=False):
    if reSampling:
        ros = RandomOverSampler(random_state=42)
        X, y = ros.fit_resample(X, y)
        print(f" After Re Sampling {Counter(y)}")
    # train test split
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1, test_size=0.2, stratify=y)
    
    # normalization for numerical columns
    sc = MinMaxScaler()
    norm_columns=[col+'_norm' for col in numeric_features]
    X_train_norm = pd.DataFrame(sc.fit_transform(X_train[numeric_features]), columns =norm_columns )
    X_test_norm = pd.DataFrame(sc.transform(X_test[numeric_features]), columns = norm_columns)
    
    # Ordinal encoding to covert categories to labels
    ordinalScalar=OrdinalEncoder()
    norm_columns=[col+'_oe' for col in categorical_features]
    X_train_oe=pd.DataFrame(ordinalScalar.fit_transform(X_train[categorical_features]), columns =norm_columns)
    X_test_oe=pd.DataFrame(ordinalScalar.transform(X_test[categorical_features]), columns =norm_columns)
    
    
    X_train_norm.index=X_train.index
    X_train_oe.index=X_train.index
    X_test_norm.index=X_test.index
    X_test_oe.index=X_test.index
    
    #keeping raw columns values as well as values after transformation
    X_train=pd.concat([X_train, X_train_norm, X_train_oe], axis=1)
    X_test=pd.concat([X_test, X_test_norm, X_test_oe], axis=1)
    
    # Label encoding to convert target variable
    labelEncoder=LabelEncoder()
    y_train=pd.DataFrame(labelEncoder.fit_transform(y_train), columns=['income'])
    y_test=pd.DataFrame(labelEncoder.transform(y_test), columns=['income'])
    
    return X_train,X_test,y_train,y_test, labelEncoder
```

```python
numeric_features = ['age', 'capital.gain', 'capital.loss', 'hours.per.week', 'fnlwgt']
categorical_features = ['race', 'sex', 'occupation', 'marital.status', 'relationship', 'native.country', 'workclass']
ordinal_features = ['education.num']
drop_features = ['education']
target_features = ["income"]

features=numeric_features+categorical_features + ordinal_features
X=censusIncome[features]
y=censusIncome['income']
```

```python
X_train,X_test,y_train,y_test, labelEncoder=data_preprocessing(X, y, numeric_features, categorical_features)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

### Feature Engineering

#### Identification of category in each column with rare occurance

```python
for feature in categorical_features:
    print('=' * 30)
    print(censusIncome[feature].value_counts().sort_values(ascending=False) / censusIncome.shape[0])
```

merge small percetage (tail) categorical values into single category.

```python
censusIncome["race"] = censusIncome["race"].map(lambda x: 'Other' if x in ['Amer-Indian-Eskimo', 'Other'] else x)
censusIncome["occupation"] = censusIncome["occupation"].map(lambda x: 'Other' if x in ['Priv-house-serv', 'Armed-Forces', 'Protective-serv'] else x)
censusIncome["marital.status"] = censusIncome["marital.status"].map(lambda x: 'Other' if x in ['Married-spouse-absent', 'Married-AF-spouse'] else x)
censusIncome["native.country"] = censusIncome["native.country"].map(lambda x: 'Other' if x not in ['United-States'] else x)
censusIncome["workclass"] = censusIncome["workclass"].map(lambda x: 'Other' if x in ['Without-pay', 'Never-worked'] else x)
```

Creating categorical variable for age and hour.per.week columns to check if they improve the results.

```python
censusIncome['age_label'] = pd.cut(censusIncome['age'], bins = [0, 25, 50, 100], labels = ['Young', 'Adult', 'Old'])
censusIncome['hours.per.week_label'] = pd.cut(censusIncome['hours.per.week'], 
                                   bins = [0, 30, 40, 100], 
                                   labels = ['Lesser Hours', 'Normal Hours', 'Extra Hours'])
```

capital.gain & capital.loss are mutually exclusive columns & sparse. we have createe another column to find the diffrence & assigned it to 2 category.

```python
censusIncome['capital.diff'] = censusIncome['capital.gain'] - censusIncome['capital.loss']
censusIncome['capital.diff'] = pd.cut(censusIncome['capital.diff'], bins = [-5000, 5000, 100000], labels = ['Minor', 'Major'])
sns.countplot(x = 'income', hue = 'capital.diff', data = censusIncome)
```
